```python
# Import the data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

crime = pd.read_excel('/Users/taylorlast/Documents/GitHub/School/Python/Advanced Statistical Programming/table-1.xls',skiprows=3,usecols = 'A:V')

crime = crime.iloc[0:20]
crime_df = pd.DataFrame(crime, columns= crime.columns)

#Change the column names
crime_df = crime_df.rename(columns = {'Violent\ncrime':'Violent crime','Violent \ncrime \nrate ':'Violent crime rate',
                 'Murder and\nnonnegligent \nmanslaughter':'Murder and nonnegligent manslaughter',
                 'Murder and \nnonnegligent \nmanslaughter \nrate ':'Murder and nonnegligent manslaughter rate',
                 'Rape\n(revised \ndefinition)':'Rape (revised)','Rape\n(revised \ndefinition) \nrate':'Rape (revised) rate',
                 'Rape\n(legacy \ndefinition)':'Rape (legacy)','Rape\n(legacy \ndefinition) \nrate':'Rape (legacy) rate',
                 'Robbery \nrate ':'Robbery rate', 'Aggravated \nassault':'Aggravated assault',
                 'Aggravated \nassault rate ':'Aggravated assault rate',
                 'Property \ncrime':'Property crime','Property \ncrime \nrate ':'Property crime rate',
                 'Burglary \nrate ':'Burglary rate','Larceny-\ntheft':'Larceny-theft','Larceny-\ntheft rate ':'Larceny-theft rate',
                 'Motor \nvehicle \ntheft':'Motor vehicle theft','Motor \nvehicle \ntheft \nrate ':'Motor vehicle theft rate'})
crime_df.columns
```

```python
#1a (Plot of violent crime rate versus time)

plt.plot(crime_df['Year'],crime_df['Violent crime rate'])
plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.title('Violent Crime Rate over time')
plt.xticks(range(1998,2017,3))
plt.style.use('fivethirtyeight')
plt.show()
```

```python
plt.plot(crime_df['Year'],crime_df['Violent crime rate'],crime_df['Year'],crime_df['Murder and nonnegligent manslaughter rate'],
        crime_df['Year'],crime_df['Rape (revised) rate'], crime_df['Year'],crime_df['Robbery rate'],
         crime_df['Year'],crime_df['Aggravated assault rate'])
plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.title('Different Types of Violent Crime Rate over time')
plt.xticks(range(1998,2017,3))
plt.legend(['Violent Crime','Murder','Rape','Robbery','Aggrevated Assault'])
plt.style.use('fivethirtyeight')
plt.show()
```

```python
plt.plot(crime_df['Year'],crime_df['Property crime rate'])
plt.xlabel('Year')
plt.ylabel('Property Crime Rate')
plt.title('Property Crime Rate over time')
plt.xticks(range(1998,2017,3))
plt.style.use('fivethirtyeight')
plt.show()
```

```python
plt.plot(crime_df['Year'],crime_df['Property crime rate'],crime_df['Year'],crime_df['Larceny-theft rate'],
        crime_df['Year'],crime_df['Burglary rate'], crime_df['Year'],crime_df['Motor vehicle theft rate'])
plt.xlabel('Year')
plt.ylabel('Property Crime Rate')
plt.title('Different Types of Property Crime Rate over time')
plt.xticks(range(1998,2017,3))
plt.legend(['Property Crime','Larceny-theft','Burglary','Motor vehicle theft'])
plt.style.use('fivethirtyeight')
plt.show()
```

```python
# Part b

# Create different data sets and load each one into a series so the histogram can stack by ICLEVEL
LO_1 = college_data['INC_PCT_{}'.format('LO')].loc[college_data['ICLEVEL']==1]
LO_2 = college_data['INC_PCT_{}'.format('LO')].loc[college_data['ICLEVEL']==2]
LO_3 = college_data['INC_PCT_{}'.format('LO')].loc[college_data['ICLEVEL']==3]
M1_1 = college_data['INC_PCT_{}'.format('M1')].loc[college_data['ICLEVEL']==1]
M1_2 = college_data['INC_PCT_{}'.format('M1')].loc[college_data['ICLEVEL']==2]
M1_3 = college_data['INC_PCT_{}'.format('M1')].loc[college_data['ICLEVEL']==3]
M2_1 = college_data['INC_PCT_{}'.format('M2')].loc[college_data['ICLEVEL']==1]
M2_2 = college_data['INC_PCT_{}'.format('M2')].loc[college_data['ICLEVEL']==2]
M2_3 = college_data['INC_PCT_{}'.format('M2')].loc[college_data['ICLEVEL']==3]
H1_1 = college_data['INC_PCT_{}'.format('H1')].loc[college_data['ICLEVEL']==1]
H1_2 = college_data['INC_PCT_{}'.format('H1')].loc[college_data['ICLEVEL']==2]
H1_3 = college_data['INC_PCT_{}'.format('H1')].loc[college_data['ICLEVEL']==3]
H2_1 = college_data['INC_PCT_{}'.format('H2')].loc[college_data['ICLEVEL']==1]
H2_2 = college_data['INC_PCT_{}'.format('H2')].loc[college_data['ICLEVEL']==2]
H2_3 = college_data['INC_PCT_{}'.format('H2')].loc[college_data['ICLEVEL']==3]

plt.figure(figsize = (25,10))
LO = [LO_1,LO_2,LO_3]
M1 = [M1_1,M1_2,M1_3]
M2 = [M2_1,M2_2,M2_3]
H1 = [H1_1,H1_2,H1_3]
H2 = [H2_1,H2_2,H2_3]

plt.subplot(151)
plt.hist(LO,stacked=True, bins = 20, alpha = .7)
plt.title("income $0-$30k")
plt.xlabel('Percentage of Students')

plt.subplot(152)
plt.hist(M1,stacked=True, bins = 20, alpha = .7)
plt.title("income $30,001-$48,000")
plt.xlabel('Percentage of Students')

plt.subplot(153)
plt.hist(M2,stacked=True, bins = 20, alpha = .7)
plt.title("income $48,001-$75,000")
plt.xlabel('Percentage of Students')

plt.subplot(154)
plt.hist(H1,stacked=True, bins = 20, alpha = .7)
plt.title("income $75,001-$110,000")
plt.xlabel('Percentage of Students')

plt.subplot(155)
plt.hist(H2,stacked=True, bins = 20, alpha = .7)
plt.title("income $110,001+")
plt.xlabel('Percentage of Students')


plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Question 1
# 1i

r = 100000
n = 10
p = .5

sim_coin = np.random.uniform(size=n*r).reshape((n,r))
sim_coin_bool = sim_coin < p
sim_coin_bool.shape
```

```python
# Convert both matricies to dataframes
sim_coin = pd.DataFrame(sim_coin)
sim_coin_bool = pd.DataFrame(sim_coin_bool)
```

```python
#1ii
sim_coin_means = sim_coin_bool.apply(np.sum,0)
sim_coin_means
```

```python
#1iii
num_of_success = pd.DataFrame(sim_coin_means.value_counts()).sort_index()
num_of_success
```

```python
#1iv
from scipy.stats import binom

real_binom = binom.pmf(range(11),n=10,p=.5) * r

real_binom = pd.DataFrame(real_binom)
real_binom
```

```python
binom_compare = pd.concat([num_of_success, real_binom],axis=1)
binom_compare.columns = ['sammple','real binomial']
binom_compare
```

```python
# Question 2
# 2a

r = 20
n = 10
meanseq = np.array(range(1,11))-1
sdseq = np.array(range(1,11))*.5

normal_mat = pd.DataFrame(np.random.normal(meanseq,sdseq,size=(r,n)))
normal_mat
```

```python
#2a - means
normal_mat_means = normal_mat.apply(np.mean,0)
normal_mat_means
```

```python
#2a - sd
normal_mat_sd = normal_mat.apply(np.std,0)
normal_mat_sd
```

```python
# 2b
normal_mat_within20 = ((normal_mat_means > meanseq*.8) & (normal_mat_means < meanseq*1.2) & (normal_mat_sd > sdseq*.8) & (normal_mat_sd < sdseq *1.2))
sum(normal_mat_within20)
```

```python
# 2c 

cols = meanseq[normal_mat_within20]
normal_mat_new = normal_mat[cols]
normal_mat_new

# Alternate way: 
# from itertools import compress
# cols = list(compress(normal_mat,normal_mat_within20))
# normal_mat[cols]
```

```python
# 2d
normal_mat_new_means = normal_mat_new.apply(np.mean,0)
normal_mat_new_sd = normal_mat_new.apply(np.std,0)

normal_mat_standardized = (normal_mat_new-normal_mat_new_means)/(normal_mat_new_sd)
normal_mat_standardized
```

**Analysis for 2a:**
Based on the bar plots in 2a we can draw conclusions about the frequency of intent of gun death for each race and gender. The first thing that sticks out is that men have many more gun deaths than women across all levels of intent and race. This could be due to men getting into conflicts more frequently than women, and it also appears that the men suicide rate is much higher than women. Based on the data, we see that Hispanics and Blacks age distribution is skewed right, which indicates that a lot of younger people die from suicide and homicide, while for the Whites age distribtuion we see a more symmetrical distribtuion (There is still a spike for Whites around 20).
